## Histograma

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv("data/trees.csv")
print(base.shape)
base
```

```python
# Cria o histograma, considerando somente o segundo atributo com 6 barras
hist = np.histogram(base.iloc[:,1], bins=6)
hist
```

```python
# Histograma com 6 barras
plt.hist(base.iloc[:,1], bins=6)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

```python
# Histograma com 10 barras
plt.hist(base.iloc[:,1], bins=10)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

## Gráficos de Densidade

```python
base = pd.read_csv("data/trees.csv")
base.head()
```

```python
# Histograma com linha de distribuição de frequência, com 6 divisões, 
# e sem linha de densidade (kde)
sns.distplot(base.iloc[:,1], hist=True, kde=False, bins=6,
    color="blue", hist_kws={'edgecolor':"black"}
)
```

```python
# Densidade 
sns.distplot(base.iloc[:,1], hist=False, kde=True, bins=6,
    color="blue", hist_kws={'edgecolor':"black"}
)
```

```python
sns.distplot(base.iloc[:,1], hist=True, kde=True, bins=6,
    color="blue", hist_kws={'edgecolor':"black"}
)
```

## Gráfico de Dispersão


```python
# Grafico de dispersão considerando o volume e a dispersão
plt.scatter(base.Girth, base.Volume, color="blue", facecolors="none", marker="*")
plt.title("Árvores")
plt.xlabel("Volume")
plt.ylabel("Circunferência")
```

```python
# Grafico de linha considerando o volume e o girth
plt.plot(base.Girth, base.Volume)
plt.title("Árvores")
plt.xlabel("Volume")
plt.ylabel("Circunferência")
```

```python
# Gráfico de dispersão com afastamento (jitter) e sem linha de tendencia (fit_reg)
sns.regplot(x=base.Girth, y=base.Volume, data=base, x_jitter=0.3, fit_reg=False)
```

```python
base = pd.read_csv("data/co2.csv")
base.head()
```

```python
x = base.conc
y = base.uptake
```

```python
# Gambiarra pra pegar os tratamentos únicos
unicos = list(set(base.Treatment))
unicos
```

```python
# Percorre os tratamentos e cria um grafico de dispersão
for treat in unicos:
    indice = base.Treatment == treat
    plt.scatter(x[indice], y[indice], label=treat)

plt.legend(loc="lower right")
```

## Gráfico de Dispersão Seaborn

```python
base = pd.read_csv("data/co2.csv")
```

```python
sns.scatterplot(x=base.conc, y=base.uptake, hue=base.Type)
```

```python
q = base.loc[base["Type"] == "Quebec"]
m = base.loc[base["Type"] == "Mississippi"]
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=q.conc, y=q.uptake).set(title="Quebec")
plt.subplot(1,2,2)
sns.scatterplot(x=m.conc, y=m.uptake).set(title="Mississippi")
plt.tight_layout()
```

```python
ch = base.loc[base["Treatment"] == "chilled"]
nc = base.loc[base["Treatment"] == "nonchilled"]
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=ch.conc, y=ch.uptake).set(title="Chilled")
plt.subplot(1,2,2)
sns.scatterplot(x=nc.conc, y=nc.uptake).set(title="Non Chilled")
plt.tight_layout()
```

## Gráfico 3D


```python
from mpl_toolkits.mplot3d import axes3d
base = pd.read_csv("data/orchard.csv")
base.head()
```

```python
fig = plt.figure()
axis = fig.add_subplot(1,1,1, projection='3d')
axis.scatter(base.decrease, base.rowpos, base.colpos)
axis.set_xlabel("decrease")
axis.set_ylabel("rowpos")
axis.set_zlabel("colpos")
```
