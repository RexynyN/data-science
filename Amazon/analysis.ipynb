```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime 

TODAY = datetime.now().strftime('%Y-%m-%d')

df = pd.read_csv("amazon_data.csv")
df.head()
```

### Preços que diminuíram

```python
for index, row in df.iterrows():
    last_price = row[-2:][0]
    latest_price = row[-2:][1]
    if latest_price < last_price:
        print(f"{row['title']} => De R${last_price} para R${latest_price}")
```

### Preços que aumentaram

```python
for index, row in df.iterrows():
    last_price = row[-2:][0]
    latest_price = row[-2:][1]
    if latest_price > last_price:
        print(f"{row['title']} => De R${last_price} para R${latest_price}")
```

```python
for index, row in df.sort_values(by=TODAY, ascending=True).iterrows():
    if row[TODAY] != -1.0:
        print(f"{row['title']} => R${row[TODAY]}")
```
