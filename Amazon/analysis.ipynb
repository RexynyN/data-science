```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime 

TODAY = datetime.now().strftime('%Y-%m-%d')

df_prices = pd.read_csv("amazon_data.csv")
df_prices.tail()
```

```python
def compare_dfs(df, func):
    for _, row in df.iterrows():
        last_price = row[-2:][0]
        latest_price = row[-2:][1]
        if func(last_price, latest_price):
            print(f"{row['title']} => De R${last_price} para R${latest_price}")
```

### Preços que diminuíram

```python
compare_dfs(df_prices, lambda x, y: x > y)
```

### Preços que aumentaram

```python
compare_dfs(df_prices, lambda x, y: x < y)
```

```python
for index, row in df_prices.sort_values(by=TODAY, ascending=True).iterrows():
    if row[TODAY] != -1.0:
        print(f"{row['title']} => R${row[TODAY]}")
```
