```python
import pandas as pd
import numpy as np

df = pd.read_csv('P2_Prices.csv')

# Sector just can't be null
df.drop(df.loc[df['Sector'].isnull()].index, inplace=True)

# Clean it all up 
cols = df.columns.tolist()[2:]
for col in cols:
    df[col].fillna(df[col].median(), inplace=True)

df.dropna(how='all', inplace=True)

df.isnull().sum()
# df.tail()
# df
    
```

```python
# Clona o dataframe para armazenar 
df_var = pd.DataFrame(columns=df.columns)
df_var[['Ticker', 'Sector']] = df[['Ticker', 'Sector']]

index = 3
for col in df.columns.tolist()[3:]: 
    df_var[col] = (df.iloc[:, index] / df.iloc[:, index - 1]) - 1
    index += 1

del df_var['6/30/2022']

df_var.head()
df_var
```

```python
df_avg = df_var.groupby(["Sector"]).mean()

sectors = df_avg.index 
df_avg.reset_index(drop=True, inplace=True)

df_avg.head()
```
