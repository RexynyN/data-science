```python
from collections import Counter 
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Create buckets of grades, considering 100 in the 90s bucket 
hist_data = Counter(min(grade // 10 * 10, 90) for grade in grades)


# Create the histogram plot 
plt.bar([x + 5 for x in hist_data.keys()], # Shift bars to appear in the right bucket
        hist_data.values(),
        10,                                # Bar width 
        edgecolor=(0, 0, 0)                # Bar edges
    )

plt.axis([
    -5, 105,  # x axis 
    0, 5      # y axis 
])

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of P1 Exam Grades");

```

```python
# The plot of the statistic malandragem (how to trick the feeble)

mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Not So Huge Anymore")
plt.show()
```

### Line Charts

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]


total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='Variance')     # Green solid line
plt.plot(xs, bias_squared, 'r-.', label='Bias^2')  # Red dot-dashed
plt.plot(xs, total_error, 'b:', label='Total Error')


# Because we've assigned labels to each series,
# we can get a legend for free (loc=9 means "top center")
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The Bias-Variance Tradeoff");
```

### Scatter Plot

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']


plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
        xy=(friend_count, minute_count),
        xytext=(5, -5),
        textcoords='offset points'
    )

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site");
```
