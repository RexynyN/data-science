```python
# Droping rows with null values
poke.dropna(how="any") # if any null value is present (default)
# or
poke.dropna(how="all") # only if all row/column values is null
```

```python
poke = pd.read_csv(join("data", "pokemon_data.csv"))

# Filling null values
poke['Speed'].fillna(poke['Speed'].mean(), inplace=True)
# or
poke.fillna(method='ffill', inplace=True) # Gets the next valid value and fill with it
```

```python
poke['Type 2'].fillna("Nothing", inplace=True) # Nothing is a place holder

# Findind and changing unique values
poke['Type 2'].value_counts()

poke['Type 2'].replace("Nothing", "-")
# or
poke['Type 2'].replace({'Nothing': '-', 'Dark': 'Umbreal'})
```

```python
# Get duplicated values
poke.duplicated(subset=['Name']).value_counts() # If all columns, leave blank

# Drop duplicates
poke.drop_duplicates(subset=['Name'], keep='last') # or First
```

```python
poke.sort_values('Name')
# or
poke.sort_values("Name", ascending=False)
# or
poke.sort_values(['Type 1', 'HP'], ascending=[1,0])
```

```python
# Add new Column
poke['Total'] = poke['HP'] + poke['Attack'] + poke['Defense'] + poke['Sp. Atk'] + poke['Sp. Def'] + poke['Speed']
poke.head()
```

```python
# Delete column
poke.drop(columns=['Total'], inplace=True)
# or
del poke['Total']
```

```python
# Other way to create a column
poke['Total'] = poke.iloc[:, 4:10].sum(axis=1)
poke.head()
```

```python
# Rearranging columns using indexes
cols = poke.columns.tolist()
df = poke[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

```python
# Save the datafram 
df.to_csv(join('data', 'modified_pokemon_data.csv'), index=False)
# or
df.to_excel(join('data', 'modified_pokemon_data.csv'), index=False)
# or 
df.to_csv(join('data', 'modified_pokemon_data.txt'), index=False, sep="\t")

```

```python
# Filtering data
df.loc[(df['Type 1'] == "Grass") & (df['Type 2'] == 'Poison')].head()
# or
df.loc[(df['Type 1'] == "Grass") | (df['Type 2'] == 'Poison')].head()
# or
new_df = df.loc[(df['Type 1'] == "Grass") & (df['Type 2'] == 'Poison') & (df['HP'] > 70)].head()
new_df
```

```python
# Reset the index of the filtered data
new_df.reset_index(inplace=True)
# or
new_df.reset_index(inplace=True, drop=True) # Drops the old index
new_df
```

```python
import re
# More filtering
df.loc[df['Name'].str.contains("Mega")].head()
# or
df.loc[~df['Name'].str.contains("Mega")].head() # not
# or
# Fire or Grass, Ignore case
df.loc[df['Type 1'].str.contains("fire|grass", regex=True, flags=re.I)].head()
```

```python
# Starts with pi (^), but has 0 or more (*) letters afterwards ([a-z])
df.loc[df['Name'].str.contains("^pi[a-z]*", regex=True, flags=re.I)].head()
```

```python
# Conditional changes
df.loc[df['Type 1'] == 'Fire', "Type 1"] = 'Flamer'
# or
df.loc[df['Name'].str.contains('Mega'), "Name"] = df['Name'].str.replace("Mega", "Hyper")
df.loc[df['Name'].str.contains('Hyper')].head()
# or 
df.loc[df['Total'] > 500, ["Generation", "Legendary"]] = 'Tester Value'
# or 
df.loc[df['Total'] > 500, ["Generation", "Legendary"]] = ['Tester 1', 'Tester 2']

df.head()
```
